```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
# read image
img = cv2.imread('facial-emotion-inputs/amit.jpg')
  
# call imshow() using plt object (to help show image)
plt.imshow(img[:, :, : : -1])
  
# display that image
plt.show()
```

```python
#analyze the face and print results

result = DeepFace.analyze(img,
                          actions = ['emotion'])
  
# print result
print(result)
```

```python
## read emotions of video frame by frame
video = cv2.VideoCapture('facial-emotion-inputs/raw-social-media-data/video.mp4')

## loop through each frame of video
while video.isOpened():
            ret, frame = video.read()
            if ret is True:
                
                ## find analysis of emotion and print
                result = DeepFace.analyze(img,
                          actions = ['emotion'])
                print(result)
            else:
                break
```
