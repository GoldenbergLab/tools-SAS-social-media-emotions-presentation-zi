```python
from fer import FER
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
img = cv2.imread("facial-emotion-inputs/amit.jpg")
detector = FER(mtcnn=True)
detector.detect_emotions(img)

faces = detector.detect_emotions(img)
print(faces)
```

```python
from fer.utils import draw_annotations

image = draw_annotations(img, faces)
plt.imshow(image)
```

```python
# Capture all the emotions on the image
captured_emotions = detector.detect_emotions(img)
# Print all captured emotions with the image
print(captured_emotions)
plt.imshow(img)
```

```python
from fer import Video

video_filename = "facial-emotion-inputs/bobo.mp4"
video = Video(video_filename)

# Analyze video 
detector = FER(mtcnn=True)
raw_data = video.analyze(detector, display=False)
df = video.to_pandas(raw_data)
```

```python
df.to_csv("bobo-emotions-by-frame.csv")
```

```python
### shows a rough graph of the emotions in the video

pltfig = df.plot(figsize=(20, 8), fontsize=16).get_figure()
```
