```python
from fer import FER
import cv2
import numpy as np
```

```python
img = cv2.imread("raw-social-media-data/amit.jpg")
detector = FER()
detector.detect_emotions(img)
```

```python
from fer import Video

video_filename = "raw-social-media-data/bobo.mp4"
video = Video(video_filename)

# Analyze video, displaying the output
detector = FER(mtcnn=True)
raw_data = video.analyze(detector, display=True)
df = video.to_pandas(raw_data)
```

```python
df.to_csv("bobo-emotions-by-frame.csv")
```

```python
pltfig = df.plot(figsize=(20, 8), fontsize=16).get_figure()
```
