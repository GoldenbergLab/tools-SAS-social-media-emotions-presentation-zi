```python
### make sure that you are in python38 (for python 3.8) environment prior to running this code! 

### ! pip3 uninstall snscrape

import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Creating list to append tweet data to
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#psychology').get_items()):
    if i>5: # small sample first
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount, tweet.retweetCount, tweet.replyCount,tweet.quoteCount])
    
# Creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username',"likeCount","retweetCount","replyCount","quoteCount"])
```

```python
# Creating list to append tweet data to
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#psychology').get_items()):
    if i>5000: # large sample of 5000 next
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount, tweet.retweetCount, tweet.replyCount,tweet.quoteCount])
    
# Creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username',"likeCount","retweetCount","replyCount","quoteCount"])
```

```python
tweets_df1.to_csv("raw-social-media-data/twitter-data-psychology-general-5000.csv")
```
